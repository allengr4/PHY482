```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect
```

```python
w=np.arange(0,5,.05)
Ratio = 1/np.sqrt(1+(1/w)**2)
plt.plot(w,Ratio)
plt.ylabel("Vin/Vout")
plt.xlabel("w")
plt.title("High-Pass Filter")
plt.show()
```

```python
t = np.arange(0,2*np.pi,.1)
Vin = 0
k = 1
while k < 10000:
    Vin=Vin+(4/np.pi)*np.sin((2*k-1)*t)/((2*k-1))
    k+=1
plt.plot(t,Vin)
plt.title("Vin")
plt.show()
```

```python
Vout = Vin/np.sqrt(2) #R=w=l=1
plt.plot(t,Vout)
plt.title("Vout")
plt.show()
```
