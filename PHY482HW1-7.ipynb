```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma
```

```python
X, Y = np.meshgrid(np.arange(-10, 10, .5), np.arange(-10, 10, .5))
I = X/np.sqrt((X**2+Y**2+.0001))**3
J = Y/np.sqrt((X**2+Y**2+.0001))**3
E = plt.quiver(X, Y, I, J, units='width')
plt.title('Point Charge at Origin')
plt.show()
```

```python
X, Y = np.meshgrid(np.arange(-1, 1, .05), np.arange(-1, 1, .05))
I2 = 5*(X/np.sqrt((X**2+(Y+.1)**2+.0001))**3 - X/np.sqrt((X**2+(Y-.1)**2+.0001))**3)#2*np.cos(Theta)*np.cos(np.sin(Theta))/np.sqrt((X**2+Y**2+.0001))**3
J2 = 5*((Y+.1)/np.sqrt((X**2+(Y+.1)**2+.0001))**3 - (Y-.1)/np.sqrt((X**2+(Y-.1)**2+.0001))**3)#2*np.cos(Theta)*np.sin(np.sin(Theta))/np.sqrt((X**2+Y**2+.0001))**3
D = plt.quiver(X, Y, I2, J2, units='width')
plt.title('Physical Dipole Charge at Origin')
plt.show()
```

```python
X, Y = np.meshgrid(np.arange(-1, 1, .05), np.arange(-1, 1, .05))
Ii = 5*((X+.1)/np.sqrt(((X+.1)**2+(Y+.1)**2+.0001))**3 - (X+.1)/np.sqrt(((X+.1)**2+(Y-.1)**2+.0001))**3)
Ji = 5*((Y+.1)/np.sqrt(((X+.1)**2+(Y+.1)**2+.0001))**3 - (Y-.1)/np.sqrt(((X+.1)**2+(Y-.1)**2+.0001))**3)
Iii = -5*((X-.1)/np.sqrt(((X-.1)**2+(Y+.1)**2+.0001))**3 - (X-.1)/np.sqrt(((X-.1)**2+(Y-.1)**2+.0001))**3)
Jii = -5*((Y+.1)/np.sqrt(((X-.1)**2+(Y+.1)**2+.0001))**3 - (Y-.1)/np.sqrt(((X-.1)**2+(Y-.1)**2+.0001))**3)
J4 = Ji+Jii
I4 = Ii+Iii
Q = plt.quiver(X, Y, I4, J4, units='width')
plt.title('Physical Quadropole Charge at Origin')
plt.show()
```
