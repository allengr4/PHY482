```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def EMF(t,B0,L,v,alpha):
    return L*v*(B0+alpha*t)
t_range = np.arange(0,5,.1)
B0_norm = 1
L_norm = 1
v_norm = 1
alpha_norm =1
plt.title('Different orders of magnitude for alpha')
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,v_norm,.1), label=".1")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,v_norm,1), label="1")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,v_norm,10), label="10")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,v_norm,100), label="100")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,v_norm,1000), label="1000")
plt.legend()
```

The behavior isn't different for larger orders of magnitude of $\alpha$, the only difference is in the size of the EMF after $t=0$.

```python
plt.title("Different Velocities")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,.1,alpha_norm), label=".1")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,1,alpha_norm), label="1")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,10,alpha_norm), label="10")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,100,alpha_norm), label="100")
plt.plot(t_range,EMF(t_range,B0_norm,L_norm,1000,alpha_norm), label="1000")
plt.legend()
```

When $v$ is altered by orders fo magnitude, $\varepsilon_{v_1}(t) \neq \varepsilon_{v_2}(t) \text{ } \forall t \in \mathbb{R}^{+}\cup\{0\} \iff v_1 \neq v_2$, whereas with $\alpha$ $\varepsilon_{\alpha_1}(0) = \varepsilon_{\alpha_2}(0)$. However, the growth is still linear with respect to time $\forall v$.
